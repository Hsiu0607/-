```python
#import/from
import cv2  #
import os  #
import random  #
import shutil  #
from sklearn.preprocessing import LabelEncoder, OneHotEncoder  #
from sklearn.utils import shuffle  #
from tensorflow.keras.models import Sequential  #
from tensorflow.keras.layers import Dense, Flatten  #
import numpy as np  #

data_path = 'C:/Users/hsiu/Desktop/efflore_peeloff'  #
train_path = 'C:/Users/hsiu/Desktop/efflore_peeloff/train'  #
test_path = 'C:/Users/hsiu/Desktop/efflore_peeloff/test'  #
class_labels = ['efflore', 'peeloff']  #
train_ratio = 0.8  #80%20%

data = []  # 
labels = []  # 

#
for i, class_label in enumerate(class_labels):  #iclass_label
    class_path = os.path.join(data_path, class_label)  #data_pathclass_label
    
    #
    for file_name in os.listdir(class_path):  #file_name
        image_path = os.path.join(class_path, file_name)  #
        
        #
        image = cv2.imread(image_path)  #image
        image = cv2.resize(image, (224, 224))  #
        
        #
        labels = np.append(labels, i)  #labels
        print('labels =', labels)  #labels

#(labels)
labels = np.array(labels)  #labelsNumPy
print('labels shape=', labels.shape)  #labels
```

```python
# 
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# 
onehot_encoder = OneHotEncoder(sparse=False)
onehot_labels = onehot_encoder.fit_transform(encoded_labels.reshape(-1, 1))

# 
print('One-hot ')
print(onehot_labels)
print(len(data))
print(len(onehot_labels))
```

```python
for i, class_label in enumerate(class_labels):
    class_path = os.path.join(data_path, class_label)
    for file_name in os.listdir(class_path):
        image_path = os.path.join(class_path, file_name)
        image = cv2.imread(image_path)
        image = cv2.resize(image, (224, 224))

        # 
        if image is None:
            print(f"Error reading image: {image_path}")
        else:
            data.append(image)

        labels = np.append(labels, i)

# dataonehot_labels
assert len(data) == len(onehot_labels)

print('Number of data samples:', len(data))
print('Number of onehot labels:', len(onehot_labels))
```

```python
from keras.utils import np_utils
import numpy as np
import tensorflow as tf
np.random.seed(10)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, onehot_labels, test_size=0.2, random_state=42)

# 
x_train_normalize = np.array(x_train) / 255.0
x_test_normalize = np.array(x_test) / 255.0

# 
x_train_flatten = x_train_normalize.reshape(x_train_normalize.shape[0], -1)
x_test_flatten = x_test_normalize.reshape(x_test_normalize.shape[0], -1)

# 
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())  # 
model.add(Dense(units=256, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
print(model.summary())
```

```python
# 
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 
train_history = model.fit(x=x_train_normalize, y=y_train, validation_split=0.2, epochs=10, batch_size=200, verbose=2)

#  x_test  NumPy 
x_test = np.array(x_test)

# 
x_test_normalize = x_test.astype('float32') / 255.0

# 
x_test_flatten = x_test_normalize.reshape(x_test_normalize.shape[0], -1)

# printx_test_normalizey_test
print('x_test_normalize shape:', x_test_normalize.shape)
print('y_test shape:', y_test.shape)
```

```python
import matplotlib.pyplot as plt
def show_train_history(train_history,train,validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title('Train History')
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.legend(['train' , 'validation'], loc='upper left')
    plt.show()
```

```python
show_train_history(train_history,'accuracy','val_accuracy')
```

```python
show_train_history(train_history,'loss','val_loss')
```

```python
scores = model.evaluate(x_test_normalize, y_test)
print('Test accuracy:', scores[1])
```
